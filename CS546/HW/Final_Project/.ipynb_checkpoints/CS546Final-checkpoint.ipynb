```python
import pandas as pd
import numpy as np
import csv
import igraph
import matplotlib.pyplot as plt
```

```python
heartdisease_data  = pd.read_csv("processed.cleveland.csv", sep= ",",names=  ["age","sex","cp","trestbps","chol",
                                                                            "fbs","restecg","thalach","exang",
                                                                            "oldpeak","slope","ca","thal","num"])
pd.DataFrame.head(heartdisease_data)
#print(heartdisease_data)
```

```python
vertices = igraph.Graph.TupleList(heartdisease_data[["age","sex","cp","trestbps","chol",
                                                    "fbs","restecg","thalach","exang",
                                                    "oldpeak","slope","ca","thal","num"]].values.tolist(), directed = True)
igraph.summary(vertices)
```
